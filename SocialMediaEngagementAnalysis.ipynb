```python
import pandas as pd
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('social_media_engagement1.csv')
```

```python
unique_platforms = df['platform'].unique()
display(unique_platforms)
```

```python
platform_counts = df['platform'].value_counts()
display(platform_counts)
```

```python
post_type_counts = df['post_type'].value_counts()
post_type_counts
```

```python
facebook_posts = df[df['platform'] == 'Facebook']
```

```python
facebook_post_type_counts = facebook_posts['post_type'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(facebook_post_type_counts, labels=facebook_post_type_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Post Types on Facebook')
plt.axis('equal')
plt.show()
```

```python
facebook_likes_by_post_type = facebook_posts.groupby('post_type')['likes'].sum()
display(facebook_likes_by_post_type)
```

```python
plt.figure(figsize=(8, 8))
plt.pie(facebook_likes_by_post_type, labels=facebook_likes_by_post_type.index, autopct='%1.1f%%', startangle=140)
plt.title('Total Likes by Post Type on Facebook')
plt.axis('equal')
plt.show()
```

```python
twitter_posts = df[df['platform'] == 'Twitter']
twitter_post_type_counts = twitter_posts['post_type'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(twitter_post_type_counts, labels=twitter_post_type_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Post Types on Twitter')
plt.axis('equal')
plt.show()
```

```python
instagram_posts = df[df['platform'] == 'Instagram']
instagram_likes_by_post_type = instagram_posts.groupby('post_type')['likes'].sum()
plt.figure(figsize=(8, 8))
plt.pie(instagram_likes_by_post_type, labels=instagram_likes_by_post_type.index, autopct='%1.1f%%', startangle=140)
plt.title('Total Likes by Post Type on Instagram')
plt.axis('equal')
plt.show()
```

```python
twitter_posts = df[df['platform'] == 'Twitter']
twitter_post_type_counts = twitter_posts['post_type'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(twitter_post_type_counts, labels=twitter_post_type_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Post Types on Twitter')
plt.axis('equal')
plt.show()
```

```python
twitter_posts = df[df['platform'] == 'Twitter']
twitter_likes_by_post_type = twitter_posts.groupby('post_type')['likes'].sum()
plt.figure(figsize=(8, 8))
plt.pie(twitter_likes_by_post_type, labels=twitter_likes_by_post_type.index, autopct='%1.1f%%', startangle=140)
plt.title('Total Likes by Post Type on Twitter')
plt.axis('equal')
plt.show()
```

```python
sentiment_by_post_type = facebook_posts.groupby(['post_type', 'sentiment_score']).size().unstack(fill_value=0)
sentiment_by_post_type['total'] = sentiment_by_post_type.sum(axis=1)
sentiment_by_post_type = sentiment_by_post_type.sort_values(by='total', ascending=False)
sentiment_by_post_type = sentiment_by_post_type.drop(columns='total')
sentiment_by_post_type.plot(kind='bar', figsize=(10, 6))
plt.title('Sentiment Distribution by Post Type on Facebook')
plt.xlabel('Number of Posts')
plt.ylabel('Post Type')
plt.tight_layout()
plt.show()
```

```python
instagram_posts = df[df['platform'] == 'Instagram']
sentiment_by_post_type_instagram = instagram_posts.groupby(['post_type', 'sentiment_score']).size().unstack(fill_value=0)
sentiment_by_post_type_instagram['total'] = sentiment_by_post_type_instagram.sum(axis=1)
sentiment_by_post_type_instagram = sentiment_by_post_type_instagram.sort_values(by='total', ascending=False)
sentiment_by_post_type_instagram = sentiment_by_post_type_instagram.drop(columns='total')
sentiment_by_post_type_instagram.plot(kind='bar', figsize=(10, 6))
plt.title('Sentiment Distribution by Post Type on Instagram')
plt.xlabel('Number of Posts')
plt.ylabel('Post Type')
plt.tight_layout()
plt.show()
```

```python
twitter_posts = df[df['platform'] == 'Twitter']
sentiment_by_post_type_twitter = twitter_posts.groupby(['post_type', 'sentiment_score']).size().unstack(fill_value=0)
sentiment_by_post_type_twitter['total'] = sentiment_by_post_type_twitter.sum(axis=1)
sentiment_by_post_type_twitter = sentiment_by_post_type_twitter.sort_values(by='total', ascending=False)
sentiment_by_post_type_twitter = sentiment_by_post_type_twitter.drop(columns='total')
sentiment_by_post_type_twitter.plot(kind='bar', figsize=(10, 6))
plt.title('Sentiment Distribution by Post Type on Twitter')
plt.xlabel('Number of Posts')
plt.ylabel('Post Type')
plt.tight_layout()
plt.show()
```

## Conclusion

Based on the charts generated:

**Post Type Distribution:**
*   On **Facebook**, carousel posts are the most frequent, followed by video and text posts.
*   On **Instagram**, carousel, text, image, and video posts have a similar distribution, with polls being the least frequent.
*   On **Twitter**, text posts are the most frequent, followed by carousel and video posts.

**Total Likes by Post Type:**
*   On **Facebook**, video posts receive the highest number of likes, followed by carousel and text posts.
*   On **Instagram**, video posts also receive the highest number of likes, followed by image and poll posts.
*   On **Twitter**, carousel posts receive the highest number of likes, followed by text and video posts.

**Sentiment Distribution by Post Type:**
*   On **Facebook**, carousel posts have a higher number of negative sentiments compared to neutral and positive. Video posts have a higher number of neutral sentiments. Text and image posts have a higher number of positive sentiments. Polls have a higher number of negative sentiments.
*   On **Instagram**, carousel and image posts have a higher number of positive sentiments. Text and video posts have a higher number of negative sentiments. Polls have a higher number of positive sentiments.
*   On **Twitter**, text, carousel and video posts have a higher number of positive sentiments. Image posts have a higher number of neutral sentiments. Polls have a higher number of positive sentiments.

Overall, video posts tend to receive a high number of likes across all platforms, even though their frequency varies. Sentiment distribution also varies significantly by platform and post type, suggesting that the type of content resonates differently with audiences on each platform.
